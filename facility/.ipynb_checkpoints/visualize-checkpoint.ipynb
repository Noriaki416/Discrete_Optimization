```python
files = glob.glob("./data/*")
data_set = 31
file_location = files[data_set]
print(file_location)
with open(file_location, 'r') as input_data_file:
    input_data = input_data_file.read()
```

```python
from collections import namedtuple
Point = namedtuple("Point", ['x', 'y'])
Facility = namedtuple("Facility", ['index', 'setup_cost', 'capacity', 'location'])
Customer = namedtuple("Customer", ['index', 'demand', 'location'])
```

```python
def parse_data(input_data):
    # parse the input
    lines = input_data.split('\n')

    parts = lines[0].split()
    facility_count = int(parts[0])
    customer_count = int(parts[1])
    print(facility_count, customer_count)
    
    facilities = []
    for i in range(1, facility_count+1):
        parts = lines[i].split()
        facilities.append(Facility(i-1, float(parts[0]), int(parts[1]), Point(float(parts[2]), float(parts[3])) ))

    customers = []
    for i in range(facility_count+1, facility_count+1+customer_count):
        parts = lines[i].split()
        customers.append(Customer(i-1-facility_count, int(parts[0]), Point(float(parts[1]), float(parts[2]))))
    return facilities, customers
```

```python
facilities, customers = parse_data(input_data)
```

```python
f_x = [f.location.x for f in facilities]
f_y = [f.location.y for f in facilities]
c_x = [c.location.x for c in customers]
c_y = [c.location.y for c in customers]
```

```python
fig = plt.figure(figsize=(5, 8),dpi=200)
ax1 = fig.add_subplot(2,1,1)
ax1.scatter(f_x,  f_y, s=5, label='f')
ax1.scatter(c_x,  c_y, s=5, label='c')
plt.legend()
```

```python
def visualize(data_set = 31):
    files = glob.glob("./data/*")
    file_location = files[data_set]
    print(file_location)
    with open(file_location, 'r') as input_data_file:
        input_data = input_data_file.read()
    
    
    facilities, customers = parse_data(input_data)
    f_x = [f.location.x for f in facilities]
    f_y = [f.location.y for f in facilities]
    c_x = [c.location.x for c in customers]
    c_y = [c.location.y for c in customers]
    
    fig = plt.figure(figsize=(5, 8),dpi=200)
    ax = fig.add_subplot(2,1,1)
    ax.scatter(f_x,  f_y, s=5, label='f')
    ax.scatter(c_x,  c_y, s=5, label='c')
    ax.set_xticks([])
    ax.set_yticks([])
    plt.legend()
    plt.show()
```

```python
visualize(4)
```
